2. Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the `BusinessTypeID` and `BusinessType` fields.

```python
# Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the BusinessTypeID and BusinessType fields
query = {"BusinessType": "Restaurant/Cafe/Canteen"}

# Return BusinessTypeID and BusinessType only
fields = {"BusinessTypeID": 1, "BusinessType": 1, "_id": 0}

# Retrieve
results = establishments.find(query, fields)

# Print
print("Matching documents:")
for doc in results:
    pprint(doc)
```

3. Update the new restaurant with the `BusinessTypeID` you found.

```python
# Update the new restaurant with the correct BusinessTypeID
# Define the query for finding the restaurant "Penang Flavours"

query = {"name": "Penang Flavours"}
update = {"$set": {"BusinessTypeID": 1}}
```

```python
# Confirm that the new restaurant was updated
updated_result = establishments.update_one(query, update)

updated_document = establishments.find_one(query)
print("Updated Document:")
pprint(updated_document)
```

4. The magazine is not interested in any establishments in Dover, so check how many documents contain the Dover Local Authority. Then, remove any establishments within the Dover Local Authority from the database, and check the number of documents to ensure they were deleted.

```python
# Find how many documents have LocalAuthorityName as "Dover"
dover_count = establishments.count_documents({"LocalAuthorityName": "Dover"})
dover_count
```

```python
# Delete all documents where LocalAuthorityName is "Dover"
delete_result = establishments.delete_many({"LocalAuthorityName": "Dover"})
```

```python
# Check if any remaining documents include Dover
print(f"Number of documents deleted: {delete_result.deleted_count}")
```

```python
# Check that other documents remain with 'find_one'
remaining_document = establishments.find_one()
pprint(remaining_document)
```

5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude

for document in establishments.find():
    # Check if the document contains 'geocode' field with 'latitude' and 'longitude'
    if 'geocode' in document and 'latitude' in document['geocode'] and 'longitude' in document['geocode']:
        
        # Convert latitude and longitude to float
        try:
            new_latitude = float(document['geocode']['latitude'])
            new_longitude = float(document['geocode']['longitude'])
        
            # Update the document with new float values
            establishments.update_one(
                {'_id': document['_id']},
                {'$set': {
                    'geocode.latitude': new_latitude,
                    'geocode.longitude': new_longitude
                }}
            )
        except ValueError:
            # Handle the exception to look for error
            print(f"Conversion error for document ID {document['_id']} with latitude {document['geocode']['latitude']} and longitude {document['geocode']['longitude']}")

print("Latitude and longitude fields have been updated to float type.")
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Change the data type from String to Integer for RatingValue
establishments.update_many(
    {"RatingValue": {"$not": {"$in": non_ratings}}},
    [{'$set': {"RatingValue": {'$toInt': "$RatingValue"}}}]
)
```

```python
# Check that the coordinates and rating value are now numbers
sample_documents = establishments.find({}, {
    '_id': 0, 
    'RatingValue': 1, 
    'geocode.latitude': 1, 
    'geocode.longitude': 1
}).limit(10)

for doc in sample_documents:
    print(doc)
```
